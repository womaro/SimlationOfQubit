This piece of code generates 1000 points on sphere. In order to ensure uniform distribution (avoid bigger condensation on polar) i use cos and arcos for z component. Following that I print points, to visualise selection.

Created figure is interactive, can be zoomed out/in and rotaded.

```python
import numpy as np
import plotly.graph_objects as go
import matplotlib.pyplot as plt

# Function to generate a single random Bloch vector
# Funkcja generująca pojedynczy losowy wektor Blocha
def generate_random_bloch_vector():
    azimuth = np.random.uniform(0, 2 * np.pi)  # Azimuthal angle (φ) / Azymutalny kąt (φ)
    z_coordinate = np.random.uniform(-1, 1)   # z-coordinate (cos(θ)) / Współrzędna z (cos(θ))
    polar_angle = np.arccos(z_coordinate)     # Polar angle (θ) / Kąt polarny (θ)

    # Convert spherical to Cartesian coordinates
    # Przekształcenie współrzędnych sferycznych na kartezjańskie
    x = np.sin(polar_angle) * np.cos(azimuth)
    y = np.sin(polar_angle) * np.sin(azimuth)
    z = z_coordinate
    return np.array([x, y, z])

# Function to generate multiple random Bloch vectors
# Funkcja generująca wiele losowych wektorów Blocha
def generate_bloch_vectors(num_vectors):
    return np.array([generate_random_bloch_vector() for _ in range(num_vectors)])

# Generate points on the Bloch sphere
# Generowanie punktów na sferze Blocha
num_points = 10000  # Number of points / Liczba punktów
bloch_vectors = generate_bloch_vectors(num_points)
x_coords, y_coords, z_coords = bloch_vectors[:, 0], bloch_vectors[:, 1], bloch_vectors[:, 2]

# Create an interactive plot
# Tworzenie interaktywnego wykresu
fig = go.Figure()

# Draw the sphere grid
# Rysowanie siatki sfery
u = np.linspace(0, 2 * np.pi, 100)  # Azimuthal angle / Kąt azymutalny
v = np.linspace(0, np.pi, 100)     # Polar angle / Kąt polarny
sphere_x = np.outer(np.cos(u), np.sin(v))  # X-coordinates of the sphere / Współrzędne X sfery
sphere_y = np.outer(np.sin(u), np.sin(v))  # Y-coordinates of the sphere / Współrzędne Y sfery
sphere_z = np.outer(np.ones(np.size(u)), np.cos(v))  # Z-coordinates of the sphere / Współrzędne Z sfery

fig.add_trace(go.Surface(
    x=sphere_x, y=sphere_y, z=sphere_z,
    colorscale="Blues",  # Sphere color / Kolor sfery
    opacity=0.5,         # Transparency of the sphere / Przezroczystość sfery
    showscale=False,
    name='Sphere / Sfera'
))

# Add points representing Bloch vectors
# Dodanie punktów reprezentujących wektory Blocha
fig.add_trace(go.Scatter3d(
    x=x_coords, y=y_coords, z=z_coords,
    mode='markers',  # Display points as markers / Wyświetlanie punktów jako znaczników
    marker=dict(size=3, color='red', opacity=0.8),  # Point styling / Stylizacja punktów
    name='Bloch Vectors / Wektory Blocha'
))

# Customize layout
# Dostosowanie układu wykresu
fig.update_layout(
    title="Bloch Vectors Distributed on a Sphere / Wektory Blocha na sferze",
    scene=dict(
        xaxis_title='X',  # Label for X-axis / Etykieta osi X
        yaxis_title='Y',  # Label for Y-axis / Etykieta osi Y
        zaxis_title='Z'   # Label for Z-axis / Etykieta osi Z
    )
)

# Show the plot
# Wyświetlenie wykresu
fig.show()
```

Now, in order to validate, whenever selection is random and uniform, I calcualte histogram, to validate uniform distribution of points.

```python
z_vals = bloch_vectors[:, 2]

# Analyze distribution
n_bins = 20
counts, bins = np.histogram(z_vals, bins=n_bins)

# Plot histogram
bin_centers = (bins[:-1] + bins[1:]) / 2
plt.bar(bin_centers, counts, width=0.1, alpha=0.7, color='orange', edgecolor='black')
plt.xlabel("z-coordinate")
plt.ylabel("Number of points")
plt.title("Distribution of Points Along z (Latitude)")
plt.show()
```

We can see, that points are generaly evenly generated on the sphere.

Now, lets choose 2 points, that will be our Lambda1 and Lambda2 vectors.

```python
# Function to select two random points from already generated Bloch vectors and highlight them
# Funkcja losująca dwa punkty z już wygenerowanych wektorów Blocha i wyróżniająca je
def highlight_two_random_points_from_existing(bloch_vectors):
    # Select two random indices from the existing points
    # Wybieramy dwa losowe indeksy z już wygenerowanych punktów
    random_indices = np.random.choice(range(len(bloch_vectors)), size=2, replace=False)
    lambda_1 = bloch_vectors[random_indices[0]]
    lambda_2 = bloch_vectors[random_indices[1]]

    # Create a new plot to highlight these two points
    # Tworzymy nowy wykres, na którym wyróżniamy te dwa punkty
    fig = go.Figure()

    # Draw the sphere grid
    # Rysowanie siatki sfery
    u = np.linspace(0, 2 * np.pi, 100)  # Azimuthal angle / Kąt azymutalny
    v = np.linspace(0, np.pi, 100)     # Polar angle / Kąt polarny
    sphere_x = np.outer(np.cos(u), np.sin(v))  # X-coordinates of the sphere / Współrzędne X sfery
    sphere_y = np.outer(np.sin(u), np.sin(v))  # Y-coordinates of the sphere / Współrzędne Y sfery
    sphere_z = np.outer(np.ones(np.size(u)), np.cos(v))  # Z-coordinates of the sphere / Współrzędne Z sfery

    fig.add_trace(go.Surface(
        x=sphere_x, y=sphere_y, z=sphere_z,
        colorscale="Blues",  # Sphere color / Kolor sfery
        opacity=0.5,         # Transparency of the sphere / Przezroczystość sfery
        showscale=False,
        name='Sphere / Sfera'
    ))

    # Add points representing Bloch vectors with transparency
    # Dodanie punktów reprezentujących wektory Blocha z przezroczystością
    x_coords, y_coords, z_coords = bloch_vectors[:, 0], bloch_vectors[:, 1], bloch_vectors[:, 2]
    fig.add_trace(go.Scatter3d(
        x=x_coords, y=y_coords, z=z_coords,
        mode='markers',  # Display points as markers / Wyświetlanie punktów jako znaczników
        marker=dict(size=3, color='red', opacity=0.1),  # Semi-transparent points / Punkty z przezroczystością
        name='Bloch Vectors / Wektory Blocha'
    ))

    # Highlight the two selected points (lambda 1 and lambda 2)
    # Wyróżnianie dwóch wybranych punktów (lambda 1 i lambda 2)
    fig.add_trace(go.Scatter3d(
        x=[lambda_1[0], lambda_2[0]], y=[lambda_1[1], lambda_2[1]], z=[lambda_1[2], lambda_2[2]],
        mode='markers+text',
        marker=dict(size=10, color=['blue', 'green'], opacity=1),  # Full opacity for highlighted points / Pełna przezroczystość dla wyróżnionych punktów
        text=["λ₁", "λ₂"],  # Use the Greek characters for the labels / Użycie greckich liter dla etykiet
        textposition="top center",  # Position of the labels / Pozycja etykiet
        name='Highlighted Points / Wyróżnione punkty'
    ))

    # Add labels for lambda 1 and lambda 2 with coordinates
    # Dodanie etykiet dla $\lambda_1$ i $\lambda_2$ z współrzędnymi
    fig.add_trace(go.Scatter3d(
        x=[lambda_1[0], lambda_2[0]], y=[lambda_1[1], lambda_2[1]], z=[lambda_1[2], lambda_2[2]],
        mode='text',
        text=[f'λ₁ ({lambda_1[0]:.2f}, {lambda_1[1]:.2f}, {lambda_1[2]:.2f})',
              f'λ₂ ({lambda_2[0]:.2f}, {lambda_2[1]:.2f}, {lambda_2[2]:.2f})'],  # Coordinates of the points / Współrzędne punktów
        textposition="bottom center",  # Position of the coordinates text / Pozycja tekstu z współrzędnymi
        showlegend=False
    ))

    # Customize layout
    # Dostosowanie układu wykresu
    fig.update_layout(
        title="Highlighted Random Points on Bloch Sphere / Wyróżnione losowe punkty na sferze Blocha",
        scene=dict(
            xaxis_title='X',  # Label for X-axis / Etykieta osi X
            yaxis_title='Y',  # Label for Y-axis / Etykieta osi Y
            zaxis_title='Z'   # Label for Z-axis / Etykieta osi Z
        )
    )

    # Show the plot
    # Wyświetlenie wykresu
    fig.show()

    return lambda_1, lambda_2  # Return the coordinates of the highlighted points / Zwrócenie współrzędnych wyróżnionych punktów

# Call the function to highlight two random points from the already generated Bloch vectors
# Wywołanie funkcji wyróżniającej dwa losowe punkty z już wygenerowanych wektorów Blocha
lambda_1, lambda_2 = highlight_two_random_points_from_existing(bloch_vectors)
```
